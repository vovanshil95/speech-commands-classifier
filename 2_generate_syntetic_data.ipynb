```python
def extract(el):
  extr = re.findall(r'\"([^\"]*)\"', el)
  if len(extr) == 0:
    extr = re.findall(r'\“([^\“]*)\”', el)
  if len(extr) == 0:
    return el
  extr = extr[0]
  if extr.startswith(' '):
    extr = extr[1:]
  return extr
```

```python
def get_audio_from_text(text: str, preset_voice: str=None, speed: float=None) -> bytes:

    request_params = {
        'text': text,
        'language_code': 'ru'
        }

    if preset_voice is not None:
        request_params['preset_voice']=preset_voice

    if speed is not None:
        request_params['speed'] = speed


    resp = requests.post(
        url='https://api.deepinfra.com/v1/inference/deepinfra/tts',
        headers={
            'Authorization': f'bearer {DEEPINFRA_TOKEN}',
            'Content-Type': 'application/json'
        },
        json=request_params
    )

    audio_content = base64.b64decode(resp.json()['audio'].split(",")[1])

    return audio_content
```

```python
def cockpit_environment(audio, sr):
    engine_noise, _ = librosa.load('585510__itinerantmonk108__mono-737-cabin-noise-at-10k-feet.wav', sr=sr)
    noise = librosa.util.normalize(engine_noise[:len(audio)]) * 0.6
    return audio + noise

def add_noise(audio_data, noise_level=0.015):
    audio, sr = librosa.load(BytesIO(audio_data), sr=16_000)
    noise = np.random.normal(0, noise_level, len(audio))

    noisy_audio = cockpit_environment(audio, sr=16_000)

    noisy_audio = noisy_audio + noise
    return noisy_audio, sr
```

```python
def generate_audio_pair(command_text, label):

    def get_gender_pair(text, label, male):
        speed = round(random.uniform(1.0, 1.5), 2)

        if male:
            voice = 'quartz'
        else:
            voice = random.choice(['luna', 'aura'])

        audio = get_audio_from_text(text, preset_voice=voice, speed=speed)
        noisy_audio, sr = add_noise(audio)
        res = {
            'audio': {'array': noisy_audio, 'sampling_rate': sr},
            'text': text,
            'label': label,
            'voice': voice,
            'speed': speed,
            'noise_level': 0.015,
            'cabin_noise_level': 0.6
        }

        return res


    results = [
        get_gender_pair(command_text, label, False),
        get_gender_pair(command_text, label, True)
    ]

    return results
```
